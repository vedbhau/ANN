Certainly! Here are the names of 8 activation functions:

1. Sigmoid: Also known as the logistic function. It squashes the output between 0 and 1.
   
2. **Tanh (Hyperbolic Tangent)**: It squashes the output between -1 and 1, offering better symmetry around 0 compared to sigmoid.

3. **ReLU (Rectified Linear Unit)**: ReLU(x) = max(0, x). It returns 0 for negative values and the input value for positive values.

4. **Leaky ReLU**: An extension of ReLU which allows a small, positive gradient for negative inputs to prevent dead neurons.

5. **ELU (Exponential Linear Unit)**: ELU(x) = x if x > 0; otherwise, ELU(x) = α * (exp(x) - 1), where α is a small positive constant.

6. **Softmax**: Used in the output layer for multi-class classification problems. It converts the output into probability distribution over multiple classes.

7. **Swish**: Swish(x) = x * sigmoid(x). Proposed as an alternative to ReLU, it is continuously differentiable and potentially performs better.

8. **PReLU (Parametric ReLU)**: Similar to Leaky ReLU but with a learnable parameter that adjusts the slope for negative inputs.

These activation functions are widely used in neural networks and deep learning architectures.

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def step(x):
    if x>=0:
        return 1
    else:
        return 0
```

```python
x = np.linspace (-5,5,10)
y = [step(i) for i in x]
print("x-values:" ,x)
print("y-values:" ,y)
plt.plot(x,y)
plt.xlabel('x values')
plt.ylabel('step(x)')
plt.title('step function')
plt.show()


# sigmoid function
x = np.linspace(-10,10,10)
y = (1/(1+np.exp(-x)))
print('X-values: ',x)
print('Y-values: ',y)
plt.plot(x,y)
plt.xlabel('X value')
plt.ylabel('Sigmoid(x)')
plt.title("Sigmoid functon")
plt.grid(True)
plt.show()


# Relu function
x = np.linspace(-10,10,10)
y = [max(0,i) for i in x]
print('X-values: ',x)
print('Y-values: ',y)
plt.plot(x,y)
plt.xlabel('X value')
plt.ylabel('ReLU(x)')
plt.title("ReLU functon")
plt.grid(True)
plt.show()



def leaky_relu(x):
    alpha = 0.1
    return np.where(x>=0,x,alpha*x)

x = np.linspace(-10,10,100)
y = leaky_relu(x)
#print('X-values: ',x)
#print('Y-values: ',y)
plt.plot(x,y)
plt.xlabel('X value')
plt.ylabel('Leaky ReLU(x)')
plt.title("Leaky ReLU functon")
plt.grid(True)
plt.show()


# tanh function
x = np.linspace(-10,10,100)
y = 2/(1+np.exp(-2*x))
plt.plot(x,y)
plt.xlabel('X value')
plt.ylabel('Tanh(x)')
plt.title("Tanh functon")
plt.grid(True)
plt.show()



def softmax(x):
    z = (np.exp(x)/np.exp(np.sum(x)))
    return z

x = np.linspace(-10,10,100)
y = softmax(x)

plt.plot(x,y)
plt.xlabel('X value')
plt.ylabel('Softmax(x)')
plt.title("Softmax functon")
plt.grid(True)
plt.show()
```
